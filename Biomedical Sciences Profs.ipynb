```python
import time
import pickle
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# scrape faculty from https://biosciences.stanford.edu/faculty/biosciences-faculty-database/
results_dict = {}
driver = webdriver.Chrome()
programs = ['Biochemistry', 'Biology', 'Biomedical Informatics', 'Biophysics', 'Cancer Biology', 'Chemical and Systems Biology', 'Developmental Biology', 'Genetics', 'Immunology', 'Microbiology and Immunology', 'Molecular and Cellular%20Physiology', 'Neurosciences', 'Stem Cell Biology and Regenerative Medicine', 'Structural Biology']
for prog in programs:
    url ='https://biosciences.stanford.edu/faculty/biosciences-faculty-database/#?program='+prog.replace(' ','%20')
    driver.get(url)
    time.sleep(5)
    listings = driver.find_elements_by_class_name("main_listing")
    for lis in listings:
        firstName = lis.find_element_by_class_name('firstName').get_attribute('innerText')
        lastName = lis.find_element_by_class_name('lastName').get_attribute('innerText')
        microResearch = lis.find_element_by_class_name('microResearch').get_attribute('innerText')
        key = firstName+lastName
        if key in results_dict:
            results_dict[key]['programs'].append(prog)
        else:
            results_dict[key] = {'firstName':firstName, 'lastName':lastName, 'programs':[prog], 'microResearch':microResearch}
with open('biofac_simple','wb+') as f:
    pickle.dump(results_dict,f)
print(results_dict)
```

```python
# add biosciences faculty to ontology and associate them with EDAM research areas
import pickle
vivoNS = onto.get_namespace("http://vivoweb.org/ontology/core")
import pdb

faculty = pickle.load(open('biofac_simple','rb'))
for indName, data in faculty.items():
    programs = [p.replace('%20','').replace(' ','') for p in data['programs']]
    indiv = vivoNS.FacultyMember(indName)
    indiv.lastName = data['firstName']
    indiv.firstName = data['lastName']
    for prog in programs:
        indiv.hasResearchArea.append(onto.search_one(iri='*'+prog+'RA'))
onto.save()
```
